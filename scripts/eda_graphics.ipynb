# Tourism Capstone Project

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

# EDA and Graphics

```python
main = pd.read_csv('../data/clean/main_clean.csv')
```

```python
main.head()
```

```python
global_arrivals = main.groupby('Year')['Total Arrivals (K)'].sum().reset_index()
```

```python
global_arrivals
```

```python
global_arrivals['Total Arrivals (K)'] = global_arrivals['Total Arrivals (K)']/1000
```

```python
fig, ax = plt.subplots(figsize=(7,7))
plt.plot('Year','Total Arrivals (K)', data = global_arrivals,
         color = 'blue')
ax.grid(axis='y',
       linestyle = 'dashed',
       alpha = 0.5)
plt.ylim(0, 2400)
plt.xlabel('Year')
plt.ylabel('Total Arrivals (M)')
plt.title('Global Tourist Arrivals by Year')
plt.xticks([2000, 2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016,
           2017, 2018, 2019, 2020, 2021, 2022], rotation=45)
plt.savefig('../images/global_arrivals.png');
```

```python
global_cruise = main.groupby('Year')['Cruise Passengers (K)'].sum().reset_index()
```

```python
global_cruise
```

```python
global_cruise['Cruise Passengers (K)'] = global_cruise['Cruise Passengers (K)']/1000
```

```python
fig, ax = plt.subplots(figsize=(7,7))
plt.plot('Year','Cruise Passengers (K)', data = global_cruise,
         color = 'blue')
ax.grid(axis='y',
       linestyle = 'dashed',
       alpha = 0.5)
plt.ylim(0, 50)
plt.xlabel('Year')
plt.ylabel('Cruise Passengers (M)')
plt.title('Global Cruise Passengers by Year')
plt.xticks([2000, 2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016,
           2017, 2018, 2019, 2020, 2021, 2022], rotation=45)
plt.savefig('../images/global_cruise.png');
```

```python
global_expenditure = main.groupby('Year')['Tourism Expenditure ($M)'].sum().reset_index()
```

```python
global_expenditure['Tourism Expenditure ($M)'] = global_expenditure['Tourism Expenditure ($M)']/1000
```

```python
fig, ax = plt.subplots(figsize=(7,7))
plt.plot('Year','Tourism Expenditure ($M)', data = global_expenditure,
         color = 'blue')
ax.grid(axis='y',
       linestyle = 'dashed',
       alpha = 0.5)
plt.ylim(0, 1400)
plt.xlabel('Year')
plt.ylabel('Tourism Expenditure ($B)')
plt.title('Global Tourism Expenditure by Year')
plt.xticks([2000, 2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016,
           2017, 2018, 2019, 2020, 2021, 2022], rotation=45)
plt.savefig('../images/global_expenditure.png');
```
